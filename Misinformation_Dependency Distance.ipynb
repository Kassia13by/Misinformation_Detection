```python
import stanza
stanza.download('zh')
```

```python
nlp = stanza.Pipeline('zh', processors='tokenize,lemma,pos,depparse', verbose=False)
```

```python
distance = []
for sent in trry.sentences:
  for word in sent.words:
    dist = abs(int(word.head)-int(word.id))
    distance.append(dist)
ave = sum(distance)/len(distance)
print(ave)
```

```python
import pandas as pd
fake = pd.read_csv('fake.csv')
```

```python
fake.head()
```

```python
dep_dist = []
distance = []

for i in range(344):
  process = nlp(fake["text"][i])
  for sent in process.sentences:
    for word in sent.words:
      dist = abs(int(word.head)-int(word.id))
      distance.append(dist)
  ave = sum(distance)/len(distance)
  dep_dist.append(ave)
  distance = []
```

```python
real = pd.read_csv('real.csv')
print(real)
```

```python
dep_dist_real = []
distance = []

for i in range(1760):
  process = nlp(real["text"][i])
  for sent in process.sentences:
    for word in sent.words:
      dist = abs(int(word.head)-int(word.id))
      distance.append(dist)
  ave = sum(distance)/len(distance)
  dep_dist_real.append(ave)
  distance = []
```

```python
dep_dist_fake = pd.Series(dep_dist)
fake.insert(8, column = "dependency_distance", value = dep_dist_fake)
```

```python
fake["dependency_distance"].head()
```

```python
dep_dist_real = pd.Series(dep_dist_real)
real.insert(8, column = "dependency_distance", value = dep_dist_real)
```

```python
Result_fake ='/content/FAKE.csv'
Result_real ='/content/REAL.csv'
fake.to_csv(Result_fake, encoding = 'utf-8',index = False)
real.to_csv(Result_real, encoding = 'utf-8',index = False)
```
